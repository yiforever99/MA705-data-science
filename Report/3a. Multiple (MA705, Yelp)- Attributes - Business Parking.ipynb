**(1) Create `state_df` and merge it with `Hour_df`**

We create a dataframe containing `state` column, and merge it with `Hour_df`.

First, we use the `spark.sql` fucntion to select columns `business_id`, `stars` and `state` from table `Hour` created before, and it returns a spark dataframe. The `toPandas()` method turns spark dataframe into a pandas dataframe. New dataframe is stored under the name `state_df`.    
Second, we use the `merge` function to join the dataframe `state_df` and the dataframe `Hour_df` on the `business_id` column. It returns a combined dataframe with 9 columns. The `business_id` column will only show once.

```
%python
#create a dataframe containing state column.
state_df = spark.sql("select business_id,state,stars from Hour").toPandas()

#merge dataframe "Hour_df" and "state_df" into one dataframe named "m_hour_df". 
import pandas as pd
m_hour_df = pd.merge(Hour_df,state_df,on="business_id")
m_hour_df
```

**(2) Create m_hour_df1 to store `state`, `stars`, `weekend`, `hour` and `status`.**

From the simple variable analysis, we know that many Starbucks will close on weekends. Therefore, we make this transformation to explore more informaiton. We create the `m_hour_df1` dataframe to store the `state`, `stars`, `weekend`, `hours` and `status` columns. The `status` is a new column we created to represent the open status on weekends.

In the first section, we use the `melt` function to reshape the dataframe `m_hour_df` and the `id_vars` parameter to identify the columns, which will remain unchanged. The `value_vars` is created to store former column labels, the `satdiff` and the `sundiff`. The values of former `satdiff` and `sundiff` columns will be combined into a new column named `value`. Then, the `rename` method is used to rename column from left side to the right side. Then, we store the new dataframe in `m_hour_df1`.

In the second section, we create a column named `status` to the `m_hour_df1`. The new column is used to identify whether the Starbucks is open or not on weekends. The `loc` attribute is used to select a specific column. We assign 1 and 0 seperatly to the criteria `>0` and `<=0` to represent the open and closed status.

```
%python
m_hour_df1= \
pd.melt(m_hour_df,
        id_vars   =['state','stars'],
        value_vars=['satdiff','sundiff'])
m_hour_df1=m_hour_df1.rename(columns={'variable':'weekend','value':'hour'}) 

m_hour_df1.loc[m_hour_df1["hour"]>0,"status"]=1
m_hour_df1.loc[m_hour_df1["hour"]<=0,"status"]=0
m_hour_df1
```

The output is a new dataframe, named `m_hour_df1`. It contains 2084 rows and 5 columns, including `state`, `stars`, `weekend`, `hour` and `status` columns. We can use this dataframe to analyze the intra-relationship among different variables. The detailed analysis will be demonstrated by visualization below.

##8. `stars`, `hours` and `state` in `business` table

We draw a heatmap to analyze the relationship among `stars`, `opening hour on weekends` and `state`.

To draw a heatmap, we first need to use the `pivot_table` method to create a pivot table based on the dataframe `m_hour_df1`. The `index` parameter is used to set `hours` as a level of the row index, and the `columns` parameter sets the `state` as the level of column index, and the `values` parameter sets the `stars` as the value. Then, we create a function by using the `lambda` and the `np.mean` operators to calculate the average stars of Starbucks with the same opening hours on weekends in the same states. Finally a dataframe with a new structure is created and stored as `m_hour_piv` 

Before using `heatmap()`, we use the `plt.figure` funtion in the `matplotlib` package. The function has the first parameter named `figsize`, which is used to set the size of the figure.The two elements inside figsize are the desired width and height of the figure in inches. Then when we draw the heatmap, it will stretch to fill the available space by the given size. Finally, we use the `sns.heatmap()` funtion to draw heatmap based on the dataframe `m_hour_piv`. The `cmap` parameter changes the color of seaborn heatmap. To add text over the heatmap, we can use the `annot` attribute. If annot is set to be `True`, the text will be written on each cell.

```
%python
m_hour_pvt =m_hour_df1\
.pivot_table(index='hour',
            columns='state',
            values='stars',
            aggfunc=lambda x: np.mean(x))


plt.clf()
plt.figure(figsize=(12, 8))
sns.heatmap(m_hour_pvt,cmap='coolwarm',annot=True)
display(plt.show())
```

In this heatmap, the zones in warm color indicate the Starbucks with higher average stars. From our graph, we can find that the the color distributions of different areas are very scattered, which means the weak correlation among the three variables. Observing the graph along y-axis direction, we can find that the lowest stars belong to Starbucks in ON which are open for around 6 to 7 hours on weekends. As the most lower stars appear on the part of the upper space and the part of the lower space, it seems that in the same states, the Starbucks with short opening hours on weekends will get lower stars, but it depends on the states. On the other hand, Starbucks in different states have different performances. For AB and PA, all the Starbucks get stars more than 3, no matter how long the opening periods are on weekends. In addition, in IL, SC and WI, the opening hours of Starbucks are more fixed at the certain length.

**(1) Dataframe with Business Parking Percentage and Average Star**

We use the `reset_index` to reset the dataframe index to default so that the `groupby()` method could be applied to group all types of business parking based on the value of the `index` column. Then, we use the `aggregate()` method on the grouped data to calculate the percentage of "True" value. In this case, due to the fact that the variables in park_rate_df only contain 0 or 1, we could consider that the `np.mean` method returns the percentages of "True" value.

```
%python
park_rate_df = busi_park.reset_index()\
   .rename(columns={'index':'states'}) \
   .groupby('states') \
   .aggregate(np.mean)
park_rate_df
```

From above code, a dataframe with five column is generated. This dataframe shows the percentages of Starbucks stores which have each type of parking in each state.

A view is created from the `business_new` view to show the average stars of Starbucks groupd by `state`. The value is orderd by `state`.

We use the `drop` statement to drop the view `avgstar` if it exists, the `select` statement selects `avg(stars)` and we use `as` to rename the column name for value of average star to `average_star`. Then, we use the `create` statement to creates view `state_starbucks`, and use `where` statement to only return information related with Starbucks stores.

```
%sql
drop table if exists avgstar;
create temporary view avgstar
as select state, avg(stars) as average_star from business_new 
group by state 
order by state
```

It returns a view named `avgstar` containing state and average. Values are grouped and ordered by `state`.

The dataframe named `avg_star` shows the average star of Starbucks of each state.   

The `spark.sql` turns all values selected from the `avgstar` view to a spark dataframe, and the `toPandas()` turns the spark dataframe to a pandas dataframe. The `set_index` is applied to set `state` column as the index of the dataframe. The new created dataframe is named as `avg_star_df`.

```
%python
avg_star_df = spark.sql('select * from avgstar').toPandas().set_index('state')
avg_star_df
```

A dataframe is created that contains the average star rating for Starbucks stores in each state. Now the `avg_star_df` and the `park_rate_df` can be concatenated to a new dataframe in order to draw graphs and analyze the relationships between star rating and each type of business parking.

The `pd.concat` is used to concate a new dataframe that contains information from both the `park_rate_df` and the `avg_star_df`. The first parameter of `pd.concat` is a list showing which dataframes are used. The `axis` parameter tells python the axis to concatenate along, with 1 meaning columns. The `join` parameter specifies to return those rows that shares the common index.

```
%python
concat = pd.concat([park_rate_df,avg_star_df],
          axis=1, 
          join='inner'
         )
concat
```

From above code, we obtained a dataframe which contains the information from `avg_star_df` and `park_rate_df`.
