## 1. Objects Selection

```
%sql
select name, count(business_id) from yelp.business
group by name
order by count(business_id) desc
```

By using SQL `select` and `group by`, we check the number of stores of every brand. In addtion, we use `order by` and `desc` to arrange the results in a descending order.   
   
From the output above, we find that Starbucks has the largest records, so we decide to focus on data of Starbucks. Starbucks ranks as the top business with 1,026 stores in `yelp` dataset.

```
%sql
select name, count(business_id) from yelp.business
where name Like "%tarbuck%"
group by name
order by count(business_id) desc
```

After grouping `business` by name, we realize that there are different spellings for the same business. For example, Starbucks can be written as "Starbucks", "starbucks", " Starkbucks Coffee" and so on. To keep data integrity, we decide to use the `like` operator to retrieve all the businesses which names are similar to Starbucks and keep all of them for further research. The coding in cmd 6 only gives the number of stores that use exactly "Starbucks" as name. After adding all stores with similar names, the number of observations becomes 1,042.

```
%sql
drop view if exists business_new;
create temporary view business_new
as select * from yelp.business
where name Like "%tarbuck%"
```

Since we only focus on Starbucks, we create a new temporary view for `business` table, which contains the only information for Starbucks. We name the new temporary view as `business_new`.

## 2. Join table   
We focus on `business` and `review` table. After doing a preliminary analysis on all the variables in these two tables, we decided to join these two tables, so that we can know more about the relation between a business and its review results. It will help us achieve our research goal.

**(1) Join Table**

```
%sql
/* Combine business and review*/
drop view if exists combine;
create temporary view combine
as select yelp.business.business_id, yelp.business.name, yelp.business.stars, yelp.business.is_open, yelp.business.city, yelp.business.hours,yelp.business.state, yelp.business.latitude, yelp.business.longitude, yelp.business.neighborhood, yelp.business.attributes, yelp.review.stars as review_stars, yelp.review.text, yelp.review.date, yelp.review.user_id, yelp.review.useful, yelp.review.review_id
from yelp.business
left join  yelp.review
on yelp.business.business_id == yelp.review.business_id
```

We use SQL code to create the temperary view named `combine` by joining the business and the review tables. The `business` table is left joined with the `review` table, setting `business_id` as the matching criteria. Outer join is not performed because our study focuses on Starbucks with applicable business information.

**(2) Filter Starbucks Records**

```
%sql
/* Create table for Starbucks */
drop table if exists starbucks;
create temporary view starbucks
as select *
from combine
where name Like "%tarbuck%"
```

Since we target only on Starbucks, we creat temporary view `starbarks` through using the SQL operator `like`.

**(8) Create `Hour` sql view and `Hour_df` Pandas dataframe.**

```
%sql
drop view if exists Hour;
create temporary view Hour
as select business_id,state, stars, split(hours.Saturday,"-")[0] as satopen, split(hours.Saturday,"-")[1] as satend,split(hours.Sunday,"-")[0] as sunopen,split(hours.Saturday,"-")[1] as sunend from business_new
```

```
%python
#assume none as close all day
Hour_df = spark.sql("select business_id,satopen,satend,sunopen,sunend from Hour").toPandas().fillna(value="0:0")
```

**(9) `Hour` data transformation.**

```
%python
# transfer string to datetime
import pandas as pd
  
Hour_df["satopen"]= pd.to_datetime(Hour_df['satopen'], format='%H:%M')
Hour_df["satend"]=pd.to_datetime(Hour_df['satend'], format='%H:%M')
Hour_df["sunopen"]=pd.to_datetime(Hour_df['sunopen'], format='%H:%M')
Hour_df["sunend"]=pd.to_datetime(Hour_df['sunend'], format='%H:%M')


#calculate time difference
import numpy as np
Hour_df['satdiff'] = Hour_df.satend-Hour_df.satopen
Hour_df['satdiff'] = Hour_df['satdiff']/np.timedelta64(1,'h')
Hour_df['sundiff'] = Hour_df.sunend-Hour_df.sunopen
Hour_df['sundiff'] = Hour_df['sundiff']/np.timedelta64(1,'h')
#calculate time difference over 24 hours
for i in range(len(Hour_df["satdiff"])):
  if Hour_df.iloc[i,5]<0:
     Hour_df.iloc[i,5]= Hour_df.iloc[i,5]+24

for i in range(len(Hour_df["sundiff"])):
  if Hour_df.iloc[i,6]<0:
     Hour_df.iloc[i,6]= Hour_df.iloc[i,6]+24

#Extract time from Datetime 
Hour_df['satopen'] =  pd.DatetimeIndex(Hour_df['satopen']).time
Hour_df['satend'] =  pd.DatetimeIndex(Hour_df['satend']).time
Hour_df['sunopen'] =  pd.DatetimeIndex(Hour_df['sunopen']).time
Hour_df['sunend'] =  pd.DatetimeIndex(Hour_df['sunend']).time

Hour_df
```

```
%python # Table for multiple variable analysis
hour_diff_df = Hour_df.loc[:,['satdiff','sundiff']]
hour_diff_df
```

```
%python
hour_diff_df = hour_diff_df.stack()\
  .rename_axis(['num','weekend'],axis=0)\
  .to_frame()\
  .reset_index()\
  .rename(columns={0:'diff'}) \
  .drop(axis=1, labels=['num'])
  
hour_diff_df
```
