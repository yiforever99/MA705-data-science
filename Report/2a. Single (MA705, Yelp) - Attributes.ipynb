##9. `hours` in `business` table   
`hours` variable contains the opening day and opening time information. We focus on Starbucks weekends' opening time.

**(1) Create `Hour` sql view and `Hour_df` Pandas dataframe.**

First, we are interested in finding the opening hours of Starbucks on Saturdays. We use `select` statement to select `hours.Saturday`, `hours` from `business new ` table.

```
%sql 
select hours.Saturday ,hours from business_new
```

The opening hours of saturdays and hours are listed above. We found that there are some null values.

Then, we create a view `Hour` to store the data of the store, state information, and we also use `split` method to split the opening hour and closing hour of each Starbucks according to the symbol `-`. And then we put them in `satopen` column and `satend` column respectively.

```
%sql
drop view if exists Hour;
create temporary view Hour
as select business_id, state, stars, split(hours.Saturday,"-")[0] as satopen, split(hours.Saturday,"-")[1] as satend, split(hours.Sunday,"-")[0] as sunopen, split(hours.Sunday,"-")[1] as sunend from business_new
```

The result is the view `Hour` that we created, which will be useful for our further analysis.

We want to check the view `Hour` using `select` statement.

```
%sql
select * from Hour
```

The view has 7 columns and 1,042 rows. We also can see that there are some null values in it. It should be notice that the values in the `satopen`, `satend`, `sunopen` and `sunend` columns are strings.

In this step, we would like to store all data other than `state` and `stars` from view `Hour` into a dataframe called `Hour_df`. And we also replace all null value with "0:0" for further analysis purpose using `.fillna()` method.

```
%python
Hour_df = spark.sql("select business_id,satopen,satend,sunopen,sunend from Hour").toPandas().fillna(value="0:0")
Hour_df
```

As we can see, the dataframe above contains opening hour and closing hour on Saturday and Sunday for each Starbucks store without null value. And the data type of those hours is `string` type.

**(2) Transform string type to datetime type**

In this step, we would like to transform the data type of hours to the datetime type using `pd.to_datetime()` function.The first parameter represents the object that will be transfered. The second parameter `format='%H:%M'` is used to define the format of the new results. The transferred data is assigned to the left of the equal sign.

```
%python
Hour_df["satopen"]= pd.to_datetime(Hour_df['satopen'], format='%H:%M')
Hour_df["satend"]=pd.to_datetime(Hour_df['satend'], format='%H:%M')
Hour_df["sunopen"]=pd.to_datetime(Hour_df['sunopen'], format='%H:%M')
Hour_df["sunend"]=pd.to_datetime(Hour_df['sunend'], format='%H:%M')
Hour_df
```

As a result, we transformed the data type of these four columns from object to datetime64. However, after data type has been transformed, the date `1900-01-01` stored in the system shows up. We will solve this issue later.

**(3) Calculate the opening period on Sat. & Sun.**

We would also like to calculate the opening period of each Starbucks store. 

To be more specific, we use `.satend` and `.satopen` to retrieve the closeing time and opening time of Starbucks on Saturday. Then we calculate the opening period of each Starbucks store using closing time minus opening time on Saturday and Sunday and put them into column `satdiff` and column `sundiff` respectively. The data type is `timedelta64`. 

In order to get a more readable result, we use `.timedelta64()` method to round the result to one decimal place with the help of the first parameter `1`, and show the opening time duration in hours using the second parameter `'h'`.

```
%python
#calculate time difference
Hour_df['satdiff'] = Hour_df.satend-Hour_df.satopen
Hour_df['satdiff'] = Hour_df['satdiff']/np.timedelta64(1,'h')

Hour_df['sundiff'] = Hour_df.sunend-Hour_df.sunopen
Hour_df['sundiff'] = Hour_df['sundiff']/np.timedelta64(1,'h')

Hour_df
```

The result above shows the opening period of each Starbucks store on weekends and the datatype is float64.

Considering that some of stores may open until the next day, we find that using the method above will cause some negative opening periods. In order to solve this problem, we can select those negative values using `if` statement, add 24 to them, and replace them with the help of `for` loop function.

In detail, the `for` loop will be run on the opening period row by row. In this statement, the `len` function returns the number of rows corresponding to the `satdiff` and `sundiff`, and the `range` function retrieves the range that loop will work on. Then, we use `iloc` attritbute to locate the column with opening hours. And `if` clause with `<0` will filter the negative value and add 24 to it.

```
%python
#calculate time difference
for i in range(len(Hour_df["satdiff"])):
  if Hour_df.iloc[i,5]<0:
     Hour_df.iloc[i,5]= Hour_df.iloc[i,5]+24


for i in range(len(Hour_df["sundiff"])):
  if Hour_df.iloc[i,6]<0:
     Hour_df.iloc[i,6]= Hour_df.iloc[i,6]+24


Hour_df
```

The result above shows the opening period of each Starbucks on Saturday and Sunday without negative values.

In this step, we would like to remove the date `1900-01-01` shown in the pervious dataframe causing by `to_datetime` method with the help of `pd.DatetimeIndex` function. 

In detail, we use `pd.DatetimeIndex` function to box data to Timestamp objects and use the attribute `time` to return the time part of the timestamps only.

```
%python
Hour_df['satopen'] =  pd.DatetimeIndex(Hour_df['satopen']).time
Hour_df['satend'] =  pd.DatetimeIndex(Hour_df['satend']).time
Hour_df['sunopen'] =  pd.DatetimeIndex(Hour_df['sunopen']).time
Hour_df['sunend'] =  pd.DatetimeIndex(Hour_df['sunend']).time
Hour_df
```

We get a good-shaped dataframe showing information about starbuck's opening status. We will also use this dataframe for further analysis.

**(4) Summariza statistics of `hours` variable**

We creat a function named `stats` by using `def` to calculate mean, median, max, min, and standard deviation of values of `satdiff` and `sundiff` columns. The function retrieves a Series with with index indicated by `index` parameter. `DataFrame()` function combines the three Series returned by the `stats()` function into one dataframe with row index indicated by `index` parameter. And the index of each Series will become column index of the dataframe.

```
%python #descriptive info
def stats(x):
  return pd.Series([x.mean(),x.median(),x.max(),x.min(),x.std()],
                  index=["Mean","Median","Max","Min","Std"])

stats_df = pd.DataFrame([stats(Hour_df.satdiff),stats(Hour_df.sundiff),stats((Hour_df.satdiff+Hour_df.sundiff)/2)],index=["sat","sun","all"])
stats_df
```

The describtive statistics derived from variable `satdiff` and `sundiff` are summarized in the dataframe above. The average opening hours of Starbucks on Saturday is 10.75 hours, and it is very close to that on Sunday which is 10.85 hours. The maximum opening hours of Starbucks on Saturday and Sunday are 21 and 22.5 and the minimum opening hours on weekends are 0. The median opening hours of Starbucks on Saturday is 14.5 and that on Sunday is 15.00. The standard deviations of opening hours of Starbucks both on Saturday and Sunday are close to 7.30. 

In conclusion, the opening hours of Starbucks on Saturady are very similar to that on Sunday.

**(5) Data visualization of `hours` variable**

We would like to use `catplot` function to draw graph with data of `satdiff` and `sundiff` columns from the dataframe`Hour_df`. The `kind` parameter is used to indicate the type of graph is a box plot. The `height` and `width` parameters are used to adjust the size of the box, and `aspect` parameter is used to adjust the aspect ratio of the whole graph.

We would also like to reshape the dataframe and use it for further analysis.

We also use `loc` attribute to extract `satdiff` and `sundiff` and put it into a new dataframe named `hour_diff_df.` We then use `stack` method to move all column indexes to row indexes. It turns out to be a series with multiple indexes, outer level is `integer` and inner level is `weekend`. The `rename_axis` method renames the row index, and `to_frame` method turns the result into a dataframe. `reset_index` method resets the index as default. `rename` method renames the column label, and `drop` method drops the column `'num'`.

```
%python 
hour_diff_df = Hour_df.loc[:,['satdiff','sundiff']].stack()\
  .rename_axis(['num','weekend'],axis=0) \
  .to_frame()\
  .reset_index() \
  .rename(columns={0:'diff'}) \
  .drop(axis=1, labels=['num'])

hour_diff_df
```

The new dataframe contains two column `weekend` and `diff` to show the opening periods of Starbucks on Saturday and Sunday.

In order to see a more clearly relationship of the opening periods, we plan to draw a catplot.

In detail, we use `catplot` function with `kind` parameter indicating the `count` value for each opening period. The x-axis represents the opening periods of Starbucks, and the y-axis shows the frequency of each kind of opening period. The `hue` parameter indicates that the two colors are differentiating the `satdiff`and `sundif` columns. The `height` parameter is used to adjust the height of the bars.

```
%python #count
sns.catplot(data=hour_diff_df, kind="count", 
            x="diff", hue="weekend", 
            height=4, aspect=13/4)
display(plt.show())
```

The graph shows the number of Starbucks within different opening periods. From the graph above, we can compare the distibution of opening periods of Starbucks on Saturday and Sunday. It is clear that above 300 Starbucks are closed on Saturday and Sunday. The distribution of opening periods is a left skewed shape with the opening periods of most Starbucks are concentrated on the range from 15.0 to 17.0 .

We would also like to see how many Starbucks are opened and closed on Saturday and Sunday.

In detail, we put the Starbucks that are opened in Saturday in `open_sat`. The statements in `[ ]` retrieve the colunms of `satdiff` without value 0 from the dataframe. The `count` method is used to count the number of records in the series. Then we use `loc` attribute to extract the value of `diff` as the number of opening Starbucks on Saturday. Following the similar steps, we can get the results we want. 

Then we use `dataframe()` function to put the result into a dataframe called `count_close`.

```
%python #count starbucks closed on weekend
open_sat = hour_diff_df[(hour_diff_df['weekend']=="satdiff") & (hour_diff_df['diff']!=0)].count().loc["diff"]
close_sat = hour_diff_df[(hour_diff_df['weekend']=="satdiff") & (hour_diff_df['diff']==0)].count().loc["diff"]

open_sun = hour_diff_df[(hour_diff_df['weekend']=="sundiff") & (hour_diff_df['diff']!=0)].count().loc["diff"]
close_sun = hour_diff_df[(hour_diff_df['weekend']=="sundiff") & (hour_diff_df['diff']==0) ].count().loc["diff"]


count_close = pd.DataFrame(data={"open_sat":open_sat,"close_sat":close_sat,"open_sun":open_sun,"close_sun":close_sun}, index=["count"])
count_close
```

As the dataframe `count_close` shown above, the number of opening Starbucks on both Saturday and Sunday is about twice of the number of closing stores.

We want to see the result above more crearly with the help of a bar chart.

In detail, `plt.clf()` is used to clean the history of the graph. The `catplot` function is used to draw a bar plot with parameter `kind` assigned with `bar`.

```
%python
plt.clf()
sns.catplot(data=count_close, kind="bar")
display(plt.show())
```

The plot above gives us a more dicrect visualizaiton and shows the same result as the `count_close` dataframe shows.

##10. `Attribute` in `business` table

**(1) Create `park_info` view for use**

We created a view called `park_info` from `business` table to extract information related with business parking analysis. `park_info` has columns `state` and `parktype` (BusinesParking from `attributes` column) and only contain rows that the value of business parking is not null.

```
%sql
drop view if exists park_info;
create temporary view park_info
as select state, attributes.BusinessParking as parktype from yelp.business
where attributes.BusinessParking is not NULL and name like '%tarbuck%'
```

The `parktype` is a dictionary which has five pairs of keys and values, indicating the type of parking and values of True or False. Although `State` is not used in the single variable analysis, it is kept for further multiple variable analysis.

**(2) Transform the data**

The `parktype` is a column containing multiple key, value pairs. In order to get the value (True or Flase) corresponding with each key, `parktype` is extracted to compose a list of string, and then converted to a list of dictionary.

The `spark.sql` converts sql table to sqark dataframe, and `toPandas` method converts the sqark dataframe to pandas dataframe. The list comprehension then extracts elements from the dataframe where the column name is `parktype` and then composed a list of string. 

In order to extract the value True or Flase from each string, we need to transform string to dictionaries. The `for` loop gets each string from `str_list`, converts data type from string to dictionary using `literal_eval` method, and inserts the dictionaries to an empty list `l2` correspondingly. Now, we have a list of dictionaries and can be used for data extraction.

```
%python
from ast import literal_eval
pkinfo_df = spark.sql('select * from park_info').toPandas()
pkinfo_df
str_list = [i for i in pkinfo_df.loc[:,'parktype']]
l2=[]
for j in range(0,len(str_list)):
  l2.insert(j,literal_eval(str_list[j]))
l2
```

After the transformation process, we created five empty lists to store values extracted from `l2` and used the five lists to compose a new dictionary.

Five empty lists with five parking types as list names are created. The for loop extracts the values  of True/False from dictionaries in `l2` and insert them to corresponding positions in the new lists. DataFrame function creates a dataframe called `busi_park` using the values from five lists created above.

```
%python
import pandas as pd
garage = []
valet = []
validated =[]
street=[]
lot=[]
for z in range(0,len(l2)):
  garage.insert(z,l2[z]['garage'])
  valet.insert(z,l2[z]['valet'])
  validated.insert(z,l2[z]['validated'])
  street.insert(z,l2[z]['street'])
  lot.insert(z,l2[z]['lot'])
state_index = [i for i in pkinfo_df.loc[:,'state']]
busi_park = pd.DataFrame(data={'garage':garage,'valet':valet,'validated':validated,'street':street,'lot':lot}, index=state_index)
busi_park
```

The `busi_park` is reformatted and assigned to a new dataframe called `busi_park_df`. The new dataframe can be used to data visualization.

The `busi_park` is reformatted using a method chain and the newly formatted dataframe is assigned to `busi_park_df`. The `stack` method moves the 5 column label to the innermost level of the row index. The `rename_axis` method renames row indexes as `states` and `parktype`. After the stack method, the result is a Series so `to_frame` method is used to convert the Series to a dataframe. When drawing graph, the x variable should be a column, not an index, so `reset_index` method is used here to reset the index to default. Now the `busi_park_df` has three columns: `states`, `parktype`, and `0`. The `states` column is a useless column caused by resetting index, so therefore, it is dropped by applying the `drop` method to make the dataframe clearer. Last, the `rename` method renames column name `0` to `TrueFalse`.

```
 %python
busi_park_df = busi_park.stack() \
  .rename_axis(['states','parktype'],axis=0) \
  .to_frame()\
  .reset_index() \
  .drop(axis=1, labels=['states'])\
  .rename(columns={0:'TrueFalse'}) 
busi_park_df
```

There are 72 of Starbucks store have garage, 3 stores provide valet, 82 stores support street parking, 351 stores have parking lot, and zero validated.

A barplot is created using `carplot` function with `kind` parameter assigned with "count". The-x axis represents the types of business parking. `hue` parameter indicates that the colors of bars are differtiated by `TrueFalse` value, which containing two type, True and False. The last two parameters, `height` and `aspect`, are used to adjust the size of bar.

```
%python
#Graph
sns.catplot(data=busi_park_df, kind="count", 
            x="parktype", hue="TrueFalse", 
            height=4, aspect=10/4)
display(plt.show())
```
