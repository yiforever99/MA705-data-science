### 2.1 `melt`

The `melt` function has two parameters `id_vars` and `value_vars`. The function: 
- creates a column named `variable` containing variable names specified with the `value_vars` parameter
- creates a column named `value` containing values from variables specified with the `value_vars` parameter
- creates a new dataframe with all columns listed by the `id_vars` parameter
- creates a row in the new dataframe for each value in each variable specified with the `value_vars` parameter

## 4. `pivot_table` with `aggfunc` parameter

```
%python
column_names = ['symboling', 'normalized-losses', 'make', 'fuel-type',
                'aspiration', 'num-of-doors', 'body-style', 'drive-wheels',
                'engine-location', 'wheel-base', 'length', 'width',
                'height', 'curb-weight', 'engine-type', 'num-of-cylinders',
                'engine-size', 'fuel-system', 'bore', 'stroke',
                'compression-ratio', 'horsepower', 'peak-rpm', 'city-mpg',
                'highway-mpg', 'price']
import_df = pd.read_csv('/dbfs/mnt/datalab-datasets/file-samples/imports-85.csv',
                        names=[string.replace('-','_') for string in column_names],
                        na_values=['?']
                       )
```

__Interpretation 13:__    
(1) Introduce the following code
- Apply a method chain on the `import_df` dataframe.   
- use **loc** method to retrieve the `body_style`, `drive_wheels` and `horsepower` columns;
- use **group by** method to group the data into 2 levels, the outside level is `body_style`, the inner level is `drive_wheels`;
- use **aggregate** method to retrieve the mean of the `horsepower`;
- use **astype** method to change the data type of values into integer.

(2) Explain the output from this cell
 - The output dataframe has multiindex,  which is composed by `body_style` and `drive_wheels`. The dataframe has only one column, named `horsepower`. The value of the column is the average horsepower defined by `body_style` and `drive_wheels`. 
 - The datatype of the value is integer.

```
%python
import_df \
  .loc[:,['body_style','drive_wheels','horsepower']] \
  .groupby(['body_style','drive_wheels']) \
  .aggregate(np.mean) \
  .astype('int')
```

__Exercise 14:__ 
- Introduce the following code cell

__Interpretation 14:__    
(1) Introduce the following code cell
- Apply a method chain on the `import_df` dataframe.   
- use **loc** method to retrieve the `body_style`, `drive_wheels` and `horsepower` columns;
- use **groupby** method to group the data into 2 levels, the outside level is body_style, the inner level is drive_wheels;
- use **aggregate** method to retrieve the mean of the horsepower;
- use **unstack** method to move the row index `drive_wheels` to the innermost level of the column index;
- use **round** method to round the value into 3 decimals.

(2) Explain the output from this cell
- The output is a dataframe. The dataframe has multiindex for columns,  which is composed by `horsepower` and `drive_wheels`. The row index is `body_style`. The value of each cell is the mean horsepower of each vehicle model defined by body_style and drive_wheels. The data is rounded to 3 decimals.

```
%python
import_df \
  .loc[:,['body_style','drive_wheels','horsepower']] \
  .groupby(['body_style','drive_wheels']) \
  .agg(np.mean) \
  .unstack('drive_wheels') \
  .round(3)
```

- The **values** parameter defines all variables in the `index` and `columns` parameters as the value of horsepower;
- The **aggfunc** parameter defines the values of each cell in the dataframe to be the mean value of horsepower. 

(2) Explain the output from this cell
- The output is a dataframe. The column label is `drive_wheels`. The row index is `body_style`. The value of each cell is the mean horsepower of each vehicle model defined by `body_style` and `drive_wheels`.

```
%python
import_df.pivot_table(index='body_style', 
                      columns='drive_wheels',
                      values='horsepower', 
                      aggfunc=np.mean)
```
