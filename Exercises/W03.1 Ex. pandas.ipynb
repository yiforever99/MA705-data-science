__9. Exercise:__ create a dataframe with:
- five rows 
- a character column called `country` containing the names of five countries
- a numeric column called `population` containing the recent populations of these countries
- a boolean column called `island` which is `True` only if the country is an island 
- at least one island country and one country that is not an island

Display the dataframe after creating it.

**Interpretation:**   
1) Code:
- **'pd.DataFrame()'** function is used to create dataframe from dictionary `data`. 
- Each element of the dictionary is a column in dataframe. Dictionary keys are columns labels.

2) Output:

```
%python import pandas as pd
country_info = pd.DataFrame(data={'country':['usa','can','aus','uk','rus'],'population': [3, 4, 5, 4, 6], 'island':[False, False, True, True, False]})
print(country_info)
```

__15. Exercise:__ create four versions of the dataframe as above (in four code cells) with a datetime index using the following parameters:
1. `start`, `end`, and `periods`
1. `start`, `end`, and `freq`
1. `start`, `periods`, and `freq`
1. `end`, `periods`, and `freq`

Each cell should store the dataframe it creates in `datetime_df`. The last line of each cell should be `(datetime_df.head(3), datetime_df.tail(3))`.

__Interpretation:__   
1) Code:
- Create a timestamp `start_datetime` and `end_datetime` using the **"to_datetime"** function. The function converts its argument to datetime, and a format argument is passed to indicate the format of the string to be converted into a datetime value.
- Create dataframe as above, and use the **"date_range"** function to return a fixed frequency DatetimeIndex. Exactly three of the four parameters `start`, `end`, `periods`, and `freq` must be specified. Three of parameters have to be used for indexes. `Periods` determines the number of period , and `freq` determines which measurement(Year,Month,Day,Hour,Minutes) to use as the interval.
- **'head(3)'** and **'tail(3)'** methods are used to show the first and last three rows of the dataframe.

2) Output:     
- Return two new dataframes. Both contains 3 rows and 2 coloumns, and have the time indexes.

```
%python
#1
start_datetime= pd.to_datetime('201901010000', format = '%Y%m%d%H%M')
end_datetime = pd.to_datetime('201907230000', format = '%Y%m%d%H%M')
datetime_df = pd.DataFrame(data= np.random.randint(0,24,size=(168,2)),
                          columns=['col_a','col_b'],
                          index = pd.date_range(start=start_datetime, end=end_datetime, periods=168))

datetime_df.head(3), datetime_df.tail(3)
```

```
%python
#2
start_datetime= pd.to_datetime('201901010000', format = '%Y%m%d%H%M')
end_datetime = pd.to_datetime('201901072300', format = '%Y%m%d%H%M')
datetime_df = pd.DataFrame(data= np.random.randint(0,24,size=(168,2)),
                          columns=['col_a','col_b'],
                          index = pd.date_range(start_datetime, end_datetime, freq = 'H'))
datetime_df.head(3), datetime_df.tail(3)
```

```
%python
#3
start_datetime= pd.to_datetime('201901010000', format = '%Y%m%d%H%M')
end_datetime = pd.to_datetime('201901072300', format = '%Y%m%d%H%M')
datetime_df = pd.DataFrame(data= np.random.randint(0,24,size=(168,2)),
                          columns=['col_a','col_b'],
                          index = pd.date_range(start_datetime, periods=168, freq = 'H'))
datetime_df.head(3), datetime_df.tail(3)
```
