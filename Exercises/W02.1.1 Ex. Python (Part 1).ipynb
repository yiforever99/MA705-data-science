## Comprehensions

__Exercise 7:__ use a list comprehension to display a list of the square of every integer greater than or equal to `0` and (strictly) less than `10`.

**interpretation**    
1)Code:
- The **range()** function generates the integer numbers between the given start integer to the stop integer, in this case from ``0`` to ``9``. 
- In the **list comprehension**, **for loop** clause is used to iterate over a sequence of numbers produced by the **range()** function and store it in variable `X`. 
- `X**2` is calculated and the corresponding values are stored in `a_list` as a list
- The **print()** function display the `a_list`.   

2)Output:   
- A list of squared values of each integer from 0 to 9

```
%python
a_list =[X**2 for X in range(10)]
print(a_list)
```

__Exercise 8:__ use a list comprehension to display a list of the square of every odd integer greater than or equal to `0` and (strictly) less than `10`.

**Interpretation**   
1) Code:  
- The **range()** function generates the integer numbers between the given start integer to the stop integer, in this case from ``0`` to ``9``. 
- The **if** statement `X %2 !=0` filters the odd numbers produced by the **range()** function.
- In the list comprehenstion,  **for loop** is used to iterate over each number produced by the **range()** and **if** function   , and store it in variable `X`.
- Then `X**2` is calculated and the corresponding values are stored in `b_list` as a list
- The **print()** function display the `b_list`, which is a list.

2) Output:   
- A list of squared values of each odd integer from 0 to 9

```
%python
b_list=[X**2 for X in range(10) if X %2 !=0]
print(b_list)
```

__Exercise 9:__ use a dictionary comprehension to create and display a dictionary with the same keys as `europe_cities` but with values that are the length of the list for that key in the `europe_cities` dictionary. For instance the value corresponding to the key `spain` will be `4`.

1) Code: 
In the dictionary comprehension,
- **Item()** method changes the key-value pairs stored in dictionary `europe_cities` as tuples;
-  Tuples are retrieved by **the `for` loop**.
-  The key and value of the resulting dictionary can be changed so that the **len()** function returns the length of each value retrieved;
-  New key value pairs ``key:len(value)`` are stored in new dictionary `europe_citiesnew`;
-  The **print()** function displays the new dictionary `europe_citiesnew`.   

2) Output:
- It returns a dictionary. 
- The keys of the dictionary are names of the country(original key of the `europe_cities` dinctionary).
- The values of the dictionary are the length of the list for the corresponding keys. The length indicates the number of the elements that each list contains.

```
%python
europe_cities = {'spain'  :['madrid', 'barcelona', 'seville', 'valencia'], 
                 'france' :['paris', 'marseille', 'bordeaux'], 
                 'germany':['berlin', 'munich'], 
                 'norway' :['oslo'] }
europe_citiesnew={key:len(value) for (key,value) in europe_cities.items()}
print(europe_citiesnew)
```

## Map & filter

__Exercise 12:__ use the `map` function to display a list of the square of every integer greater than or equal to `0` and (strictly) less than `10`.

**Interpretation:**

1) code
- **lambda** creates an anonymous function to store `x**2` in the variable `x`
- The **range()** function generates the integer numbers between the given start integer to the stop integer, in this case from ``0`` to ``9``
- **map()** function applies the **lambda** function to the variables returned from **range()** function
- **list()** function converts iterable value returned by **map()** function to a list 

2) output
- A list of the square of every integer from `0` to `9`.

```
%python
list(map(lambda x: x**2,
         range(10)))
```

__Exercise 13:__ use the `map` and `filter` functions to display a list of the square of every odd integer greater than `0` and (strictly) less than `10`.

**Interpretation:**   
1) Code
- **filter()** function returns only those elements from 0 to 9 for which the **lambda** function `x%2 !=0` returns `True`
- **map()** function runs the function defined by **lambda** to each elements in `odd`. 
- **list()** function converts iterable value returned by **map()** function to a list

2) Output
- A list of the square of every odd integer from `0` to `9`.

```
%python
odd = filter(lambda x: x%2!=0, range(10))
list(map(lambda y: y**2, odd))
```
