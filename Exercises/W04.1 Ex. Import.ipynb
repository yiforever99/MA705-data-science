__Exercise 6:__ use the `spark.sql` command, the `toPandas` method, the `loc` method, the `value_counts` method, and the `sort_values` method to display the unique values of the `stars` column from one of the Yelp tables

**Interpretation 6:**   
1) Code: 
- **spark.sql** command combined with **select** statement retrieves `stars` column from the yelp.business table and returns a Spark DataFrame.
- **toPandas** method transform the selected data to a pandas DataFrame. 
- **loc[:,'stars']** method retrieves the `stars` column and returns a Series.
- **value_counts()** counts the total number of each level of ratings as a Series.
- **sort_values** method sort the retrieved table ascendingly by the number of ratings in a Series.

2) Output:
- A Series with two columns sorted by ratings. 
- One shows rating contained in `stars` column from `yelp.business` table and another shows number of each level of ratings

```
%python
spark.sql("select stars from yelp.business") \
  .toPandas() \
  .loc[:,'stars'] \
  .value_counts() \
  .sort_values()
```

__Exercise 7:__ read one of the U.S. weather stations CSV files into Python:
1. List only rows/days with above average precipitation (include all columns in the listing)
1. List only rows/days with above average precipitation, but include only the precipitation columns

**Interpretation 7(1):**   
1)Code:
- **pd.read_csv** retrieves the csv documents column from one of the weather CSV files and store the result in `kclt_pdf`.
- **kclt_pdf['actual_precipitation']>kclt_pdf['average_precipitation']** will return a boolean Series.
- Use **kclt_pdf.loc[ ]** attribute to access specific part of dataframe, and will only returns the rows when the first index returns True.   

2)Output:
- kclt_pdf is a dataframe. 
- Print the part of dataframe only with rows above average precipitation.

```
%python
import pandas as pd
kclt_pdf = pd.read_csv('/dbfs/mnt/datalab-datasets/538_data/us-weather-history/KCLT.csv')
#1
print(kclt_pdf.loc[kclt_pdf['actual_precipitation']>kclt_pdf['average_precipitation'],:])

```

**Interpretation 7(2)**   
1) Code:
- Use **kclt_pdf.loc[ ]** attribute to access specific part of dataframe according to the first index and the second index.   
- First index for rows: **kclt_pdf['actual_precipitation']>kclt_pdf['average_precipitation']** will return a boolean Series;
- Second index for columns: **['actual_precipitation','average_precipitation','record_precipitation']** set the conditional with the specified column labels for **loc** attribute;



2) Output:
- The criteria returns the `actual_precipitation`,`average_precipitation` and `record_precipitation` columns which contains the data that is above average precipitation as a dataframe.

```
%python
#2
print(kclt_pdf.loc[kclt_pdf['actual_precipitation']>kclt_pdf['average_precipitation'],
                   ['actual_precipitation','average_precipitation','record_precipitation']])
```

__Exercise 12:__ from `per_diem_pdf` retrieve only rows with a country name of "CHINA":

**Interpretation 12:**
