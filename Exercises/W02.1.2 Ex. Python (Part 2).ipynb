__Exercise 22__: Using this file: `/dbfs/mnt/datalab-datasets/file-samples/one_list.json`
1. Create, using a `with`-`open` context manager, a list of dictionaries in variable `one_list` (from the above file)
1. Display, using the `map` function, a list of quantities (from `one_list`)
1. Display, using the `map` function, a list of quantities with `100` added to each quantity (from `one_list`)
1. Display, using the `map` and `filter` functions, a list of quantities greater than `250` (from `one_list`)

**interpretation**

1) Code:
- the **with** command opens the file for reading and stores the file object, returned by the **open** function, in the `infile` variable. 
- **json.load** function loads the contents of the file into a python list and **print()** is used to display the results

2) Output:
- A list of dictionaries in variable `one_list`

```
%python
#1
import json
with open('/dbfs/mnt/datalab-datasets/file-samples/one_list.json', 'r') as infile:
  one_list=json.load(infile)

one_list
```

**Interpretation:**

1) Code: 
- **lambda** creates an anonymous function. x in the lambda function indicates each element(subdictionary) of the `one_list`.
- **map()** function applies the lambda function, extracting the corresponding value of `'quantities'` from each dictionary.
- **list()** function converts iterable value returned by **map()** function to a list 

2) Output:
-  a list of quantities (from `one_list`)

```
%python
#2.Display, using the map function, a list of quantities (from one_list)
list_2 = list(map(lambda x:x['quantity'],one_list))
list_2
```

**Interpretation:**

1) Code:
- **lambda** creates an anonymous function. x in the lambda function indicates each element(subdictionary) of the one_list.
- **map()** function applies the lambda function, extracting the corresponding value of `'quantities'` from each dictionary.
- Add `100` to the extracted values respectively
- **list()** convert iterable value returned by **map()** function to a list 

2) output
- a list of quantities with `100` added to each quantity (from `one_list`)

```
%python
#3.Display, using the map function, a list of quantities with 100 added to each quantity (from one_list)
list_3 = list(map(lambda x:x['quantity']+100,one_list))

print(list_3)
```

**Interpretation:**

1) Code:
- **lambda** creates an anonymous function. x and y in the lambda functions indicate each element(subdictionary) of the `one_list`.
- **filter()** function applies the lambda function, indexing the corresponding value of `'quantities'` which are greater than `250` from each dictionary.The **filter** function returns the index of each target `'quantities'`
- **map()** function applies the **lambda** function, extracting the corresponding values of `'quantities'` from each dictionary, based on the index returned from the **filter()** function.
- **list()** function converts iterable value returned by **map()** function to a list 

2) Output:
- a list of quantities greater than `250` (from `one_list`)

```
%python
import json
with open('/dbfs/mnt/datalab-datasets/file-samples/one_list.json', 'r') as infile:
  one_list=json.load(infile)

criteria = filter(lambda x:x['quantity']>250, one_list)

list_4 = list(map(lambda y:y['quantity'],criteria))

list_4
```
