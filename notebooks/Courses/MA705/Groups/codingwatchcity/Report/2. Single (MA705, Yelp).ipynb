##1.`stars` in `business` table

**(1) Summarize statistics of `stars`**

First, we would like to get the basic understanding of the descriptive statistics of final star ratings such as the average, median, minimum, maximum and standard deviation of Starbucks stores. 

To be more specific，we use `.stars` to filter only the `stars` column from `busi_df` dataframe. Second, we use `mean()`, .`median()`, `min()`, `max()`, `std()`, `count()` methods to retrieve the basic statistic information of the `star` column. Then, we use the `round(2)` to round the results to two decimal places. After that, `[ ]` is used to put all the information into a list. Last, we use `pd.Series` to transform the list to a pandas series with index as "Mean, Median, Max, Min, Std and Totals".

```
%python
busi_stars_mean=busi_df.stars.mean().round(2)
busi_stars_median=busi_df.stars.median().round(2)
busi_stars_min=busi_df.stars.min().round(2)
busi_stars_max=busi_df.stars.max().round(2)
busi_stars_std=busi_df.stars.std().round(2)
busi_stars_count=busi_df.stars.count()
busi_stars_summary_list=[busi_stars_mean,busi_stars_median,busi_stars_min,busi_stars_max,busi_stars_std,busi_stars_count]
busi_stars_summary_series=pd.Series(busi_stars_summary_list,index=["Mean","Median","Min","Max","Std","Totals"])
busi_stars_summary_series
```

The output is a pandas series containing six elements. As we can see, the star rating ranges from 1 to 5 with standard deviation of 0.67, and the median star rating is 3.5.

**(2) Count the number of each rating of `stars` **

We are also interested in finding how many Starbucks stores are in each star rating using the `groupby()` method. 

In detail，we use `groupby` to group the `busi_df` dataframe by `stars` column. Second, we use `.stars` to retrieve only the `stars` column, and the result of this step is a series. Then, we use the `count()` method to count the number of different stores in each star rating. Third, we use the `to_frame()` method to transform the series to a dataframe. Last, we use the `rename()` method to change column label from `stars` to `totals`.

```
%python
count_busi_stars = busi_df.groupby(['stars']).stars.count()\
                   .to_frame()\
                  .rename(columns={'stars':'totals'})
count_busi_stars
```

The output is a dataframe, containing one column and nine rows with index name `stars` and column label `totals`. The star ratings are ranging from 1.0 to 5.0, with the step of 0.5. Most stores have the star rating between 2.5 to 4.0, and 3.5 star rating has the most 301 stores.

**(3) Plot the barcharts of `stars` by each rating**

In order to see the distribution more clearly, we would like to put it into a bar chart using `plot.bar()`method
with `count_busi_stars` dataframe. Then, the `plt.show()` and `display()` commands in `matplotlib.pyplot` package are used to show the barplots.

```
%python
count_busi_stars.plot.bar()
display(plt.show())
```

The output is a bar chart with each bar representing the number of each star rating. It turns out to be a normal distribution with most of stores having star rating between 2.5 to 4.0.

##3.`is_open` in `business` table

We would like to find out how many stores are still open using the `select` statement from `yelp.business`.
In detail, we use `select` to extract specific columns from `business_new`. Subselections is used to select columns with different criterias and the new names of selected columns are assigned followed by `AS`. Finally, we will get a summary table with three columns, showing the number of the opening Starbucks, closed Starbucks, and total Starbucks recored in the database. The `distinct` is used to show the unique results. The `is_open` variable shows whether the Starbuck is still opening with integer 0 or 1 representing for closed or open, respectively.

```
%sql
select distinct
      (select count(is_open) from business_new
      where is_open ==1 ) AS opening,
      (select count(is_open) from business_new
      where is_open ==0) AS closed,
      (select  count(is_open) from business_new) AS total
from business_new
```

It turns out that 964 stores are open and 78 stores are closed.

We would like to use `round()` function to calculate the percentage of stores that are still open.

```
%python
round(964/1042,4)
```

In our analysis, we found that around 92.51% of total Starbucks stores are in open status, which makes the analysis more practical. We have to keep in mind that a small subset of the overall population is closed already, and it may be helpful to explain the existance of possible outliers for our further analysis.

In detail,`plt.clf()` function clears the current figure so that no duplicated figues will exist. `sns.catplot()` creates a categorical data plot with data from `busi_state_df`, the x-axis shows `state` data, and the y-axis shows the count number of each state in `busi_state_df`. **(3) Create `state_starbucks` view to store the number of Starbucks in each state**

We would like to create a view `state_starbucks` to store the number of Starbucks in each state for further analysis.

In order to do this, we use `drop` statement to drop the view `state_starbucks` if it exists. Then, `select` statement is used to select `count(state)` from `business_new` and rename it as `count_state`. The `create` statement is used to create the view `state_starbucks` containing data selected above.

```
%sql
drop view if exists state_starbucks;
create temporary view state_starbucks as 
select count(state) as count_state
from business_new
group by state
```

It returns a view named `state_starbucks` containing `count(state)` from `business_new` table.

**(4) Summarize statistics of `state` variable**

We would like to get the basic understanding of the descriptive statistics of the number of Starbucks in each state such as the average,median, minimum, maximum and standard deviation. It is easy to compare them by putting them into a `pd.DataFrame`.

In detail, we use `spark` statement to create a spark dataframe with data selected from `state_starbucks`, and `toPandas()` turns it into pandas dataframe.  `count_state` statement selects the column `count_state` from pandas dataframe.  `mean()`, `min()`, `max()`, `std()` methods are applied to calculate the mean, minimum, maximum and standard deviation of `count_state`.  `pd.DataFrame()` creates a dataframe with data calculated above. The row index of this dataframe is named as `mean','min','max','std'`and the column is named as `num_starbuck_states`.

```
%python
pd.DataFrame(data=[[spark.sql('select * from state_starbucks').toPandas().count_state.mean()],[spark.sql('select * from state_starbucks').toPandas().count_state.min()],[spark.sql('select * from state_starbucks').toPandas().count_state.max()],[spark.sql('select * from state_starbucks').toPandas().count_state.std()]],index=['mean','min','max','std'],columns=['num_starbuck_states'])
```

It returns a dataframe containing mean, minimum, maximum, and standard deviation of the number of Starbucks in each state. As we can see, the number of Starbucks in each state is ranging from 2 to 256, and the average number of Starbucks is 94.727.
